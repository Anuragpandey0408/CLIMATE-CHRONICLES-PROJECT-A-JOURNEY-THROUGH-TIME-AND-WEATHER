```python
import pandas as pd

# CSV file ko load karna
file_path = 'climate_change_upto_2000_1.csv'
df = pd.read_csv(file_path)
```

```python
# "Period" column ko split karna
df[['Period_Start', 'Period_End']] = df['Period'].str.split('-', expand=True)
```

```python
# Split columns ko integer type mein convert karna
df['Period_Start'] = df['Period_Start'].astype(int)
df['Period_End'] = df['Period_End'].astype(int)
```

```python
# DataFrame ko dekhna
print(df.head())
```

matplotlib be using

```python
import matplotlib.pyplot as plt
# Data ko visualize karna
plt.figure(figsize=(14, 7))

```

```python
# Mean Temperature Maximum ka line plot
plt.subplot(3, 1, 1)
plt.plot(df['Month'], df['Mean Temperature in degree C - Maximum'], label='Max Temp', color='r')
plt.title('Mean Temperature in degree C - Maximum')
plt.ylabel('Temperature (C)')
plt.legend()
```

```python
# Mean Temperature Minimum ka line plot
plt.subplot(3, 1, 2)
plt.plot(df['Month'], df['Mean Temperature in degree C - Minimum'], label='Min Temp', color='b')
plt.title('Mean Temperature in degree C - Minimum')
plt.ylabel('Temperature (C)')
plt.legend()
```

```python
# Mean Rainfall ka line plot
plt.subplot(3, 1, 3)
plt.plot(df['Month'], df['Mean Rainfall in mm'], label='Rainfall', color='g')
plt.title('Mean Rainfall in mm')
plt.ylabel('Rainfall (mm)')
plt.legend()
```
